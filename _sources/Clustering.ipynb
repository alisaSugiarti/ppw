# **Clustering**

Clustering adalah teknik dalam analisis data dan pembelajaran mesin yang digunakan untuk mengelompokkan objek atau data ke dalam kelompok-kelompok yang memiliki kesamaan tertentu. Tujuan utama dari clustering adalah mengidentifikasi struktur dalam data dan membentuk kelompok berdasarkan karakteristik atau propertinya.
